```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Variando el número de genes permitidos por delecionar

```python
plt.style.use('bmh')
rxns = {'Threonine':'Treonina [mmol/hr]', 'Biomass':'Biomasa [hr$^{-1}$]'}
```

```python
ngenes = pd.read_csv('OptKnock_Ngenes.csv')
ngenes['Objective'] = ngenes['Objective'].replace(rxns)
ngenes.rename(columns = {'allowknock':'# de genes', 'Objective':'Función objetivo', 'Value':'Flux'}, inplace = True)
ngenes
```

```python
sns.catplot(x='# de genes', y='Flux', hue='Función objetivo', data=ngenes, kind='bar')
plt.savefig('OptKnock_Ngenes.png', bbox_inches='tight')
plt.show()
```

## Variando el mínimo de biomasa requerida

```python
minbiomass = pd.read_csv('OptKnock_MinBiomass.csv')
minbiomass['Objective'] = minbiomass['Objective'].replace(rxns)
minbiomass.rename(columns = {'minbiomass':'Mínimo de biomasa', 'Objective':'Función objetivo', 'Value':'Flux'}, inplace = True)
minbiomass
```

```python
sns.catplot(x='Mínimo de biomasa', y='Flux', hue='Función objetivo', data=minbiomass, kind='bar')
plt.savefig('OptKnock_MinBiomass.png', bbox_inches='tight')
plt.show()
```
